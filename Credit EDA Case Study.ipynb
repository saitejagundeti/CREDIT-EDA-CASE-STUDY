### Binning & Imbalance Percentage

```python
# Bins for credit amount
bins=[0,150000,200000,250000,300000,350000,400000,450000,500000,550000,600000,650000,700000,750000,800000,850000,900000,950000,1000000]
slots=['0-150000','150000-200000','200000-250000','250000-300000','300000-350000','350000-400000','400000-450000','450000-500000','500000-550000','550000-600000','600000-650000','650000-700000','700000-750000','750000-800000','800000-850000','850000-900000','900000-950000','950000 and above']
df['CREDIT_RANGE']=pd.cut(df['AMT_CREDIT'],bins,labels=slots)
```

```python
# Bins for income amount
bins = [0,25000,50000,75000,100000,125000,150000,175000,200000,225000,250000,275000,300000,325000,350000,375000,400000,425000,450000,475000,500000,10000000000]
slot = ['0-25000', '25000-50000','50000-75000','75000,100000','100000-125000', '125000-150000', '150000-175000','175000-200000',
       '200000-225000','225000-250000','250000-275000','275000-300000','300000-325000','325000-350000','350000-375000',
       '375000-400000','400000-425000','425000-450000','450000-475000','475000-500000','500000 and above']
df['INCOME_RANGE']=pd.cut(df['AMT_INCOME_TOTAL'],bins,labels=slot)
```

```python
#Creating reusable function
def func(df,col,title,hue=None):
    sns.set_style('whitegrid')
    sns.set_context('paper')
    T=pd.Series(data=hue)
    fig,ax=plt.subplots()
    plt.yscale('symlog')
    plt.title(title)
    ax=sns.countplot(data=df,x=col,hue=hue)
    plt.setp(ax.get_xticklabels(),rotation=45,ha='right',rotation_mode='anchor')
    plt.show()
```
